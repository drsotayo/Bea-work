```python
import matplotlib.pyplot as plt
```

```python
fig,ax = plt.subplots()
ax.plot(player_stats["player"], player_stats["Height"], color='b')
ax.set_xlabel("Player (Name)")
ax.set_ylabel("Height (m)")
plt.show()
```
